**Ejercicio 3**
I y J son dos variables. Input permite que el usuario les asigne un valor según indica el enunciado entre comillas y entre paréntesis. Int permite que esos datos sean transformados en valores numéricos.
If es una sentencia condicional que expresa "si se cumple X..." y el print le indica al programa que si se cumple la condición anterior genere el texto que le sigue entre comillas y entre paréntesis. Elif es otro If o condición. Else puede entenderse como "sino".

```python
i = int(input("ingresa el primer número  "))
j = int(input("ingresa el segundo número  "))
if i+j < 200:
  print("la suma de los números es menor a 200")
elif i+j > 250:
  print("la suma de los números es mayor a 250")
else:
  print("la suma está entre 200 y 250")
```

**Ejercicio 4**
La función Input permite que el usuario ingrese datos y lo que va entre paréntesis luego de ella le permite a este saber qué tipo de dato ingresar.
La función Int transforma el texto que representa un número en un valor numérico.
If es una sentencia condicional que permite darle una instrucción específica al programa si se cumplen ciertas condiciones.
Print es una función que permite que el programa escriba un texto mientras se ejecuta.
Else es otra parte de la sentencia condicional IF que agrupa todas las otras posibilidades que no cumplen las condiciones de los If.

```python
edad = input("ingresa tu edad  ")
edad = int(edad)
LeGusta= input("¿Te gusta la música urbana?  ")
if edad>=18 and LeGusta=="si":
  print("eres mayor de edad y te gusta la música urbana")
elif edad>=18 and LeGusta=="no":
  print("eres mayor de edad y no te gusta la música urbana")
else:
  print("eres menor de edad o no te gusta la música urbana")

```
